# Opening RAW images

This is a notebook with examples of opening RAW images with open_utils.py

```python
import matplotlib.pyplot as plt
import numpy as np

from open_utils import *
```

```python
directory='./imgs4'

img_indices = [1, 4, 6]
up = 0
down = 2000
left = 0
right = 2000
```

## Opening RAW images as demosaiced images

```python
imgs, raw_info = open_burst_of_frames(
    directory=directory,
    img_indices = img_indices,
    up=up, left=left,
    down=down, right=right,
    raw_needed=False
)
```

```python
plt.imshow(imgs[1])
```

## Opening RAW images as RAW images

```python
imgs, raw_info = open_burst_of_frames(
    directory=directory,
    img_indices = img_indices,
    up=up, left=left,
    down=down, right=right,
    raw_needed=True
)
```

```python
plt.imshow(imgs[1] / imgs[1].max())
```
